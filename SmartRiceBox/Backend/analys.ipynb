```python
import pandas as pd
```

```python
# Read data

rice_box_df = pd.read_csv("data/rice_box.csv")
rice_box_df
```

```python
visualize_df = pd.read_csv("data/visualize.csv")
visualize_df
```

```python
# Rename column "riebox_id" to "id"

rice_box_df = rice_box_df.rename(columns={'id': 'ricebox_id'})
rice_box_df
```

```python
# Join 2 df by id

merge_df = pd.merge(rice_box_df, visualize_df, on='ricebox_id')
merge_df
```

```python
merge_df = merge_df[['district', 'quantity', 'created_at', 'rice_type_id']]
merge_df
```

```python
# Covert created_at to month, example:

merge_df['created_at'] = pd.to_datetime(merge_df['created_at']).dt.to_period('M')
merge_df
```

```python
# Filter data by created_at from 2024-01 to 2024-08

merge_df_filtered = merge_df[(merge_df['created_at'] >= '2024-1') & (merge_df['created_at'] <= '2024-5')]
merge_df_filtered
```

```python
# Export to csv

merge_df_filtered.to_csv('data.csv', index=False)
```

```python
# Covert to list of dict with key is column

merge_df_filtered_list = merge_df_filtered.to_dict(orient='records')
merge_df_filtered_list
```
